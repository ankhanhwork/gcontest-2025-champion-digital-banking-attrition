## 1. Business Understanding

The project examined more than 26,000 onboarding attempts from 9,148 customers in Q1 2024. The objective was to identify friction in the onboarding funnel, predict final customer drop-off, and recommend measurable product improvements.

Key questions:

1. Where do customers fail or abandon the journey?
2. Which customer, behavioral, temporal, and device factors are associated with drop-off?
3. Can a model rank customers by final-attempt drop-off risk?
4. Which interventions should the bank prioritize?


```python
from pathlib import Path
import warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder

warnings.filterwarnings("ignore")
sns.set_theme(style="whitegrid", context="notebook")
RANDOM_STATE = 42
DATA_PATH = Path("../data/onboarding.csv")
```

## 2. Data Access And Confidentiality

The raw dataset is intentionally excluded. This guard prevents the notebook from silently using a different local file and makes the portfolio boundary explicit.


```python
if DATA_PATH.exists():
    raw = pd.read_csv(DATA_PATH)
    print(f"Loaded {len(raw):,} attempts with {raw.shape[1]} columns")
else:
    raw = None
    print("Private dataset not found. Documentation and reported-result cells remain available.")
```

## 3. Data Quality And Preparation

The original workflow standardized fields, parsed timestamps, ordered events within each customer, repaired stable attributes from that customer's own history, and checked missingness and implausible ages. Population-level imputation was used only after customer-consistent evidence was exhausted.


```python
COLUMN_RENAMES = {
    "cus_id": "customer_id",
    "onboarding_status": "status",
    "device_type": "phone_type",
    "drop_off_point": "stop_point",
}


def prepare_attempts(df: pd.DataFrame) -> pd.DataFrame:
    """Clean attempt-level records without using future outcome information."""
    clean = df.rename(columns=COLUMN_RENAMES).copy()
    clean.columns = clean.columns.str.strip().str.lower()

    for col in ["begin_time", "close_time"]:
        if col in clean:
            clean[col] = pd.to_datetime(clean[col], errors="coerce")

    if "status" in clean:
        clean["status"] = clean["status"].str.lower().replace({"successful": "succeed"})

    clean = clean.sort_values(["customer_id", "begin_time"]).reset_index(drop=True)

    stable_cols = [c for c in ["gender", "birth_year", "phone_type"] if c in clean]
    for col in stable_cols:
        clean[col] = clean.groupby("customer_id")[col].transform(lambda s: s.ffill().bfill())

    if {"begin_time", "close_time"}.issubset(clean.columns):
        clean["duration_seconds"] = (
            clean["close_time"] - clean["begin_time"]
        ).dt.total_seconds()

    return clean

attempts = prepare_attempts(raw) if raw is not None else None
```

## 4. Funnel And Cohort EDA

EDA was performed at both attempt and customer levels. This distinction matters: repeated failed attempts should not be counted as separate customers when estimating final drop-off.


```python
def customer_summary(df: pd.DataFrame) -> pd.DataFrame:
    ordered = df.sort_values(["customer_id", "begin_time"])
    final = ordered.groupby("customer_id", as_index=False).tail(1).copy()
    ever_succeeded = ordered.groupby("customer_id")["status"].apply(
        lambda s: s.eq("succeed").any()
    )
    final["drop_off"] = (~final["customer_id"].map(ever_succeeded)).astype(int)
    return final

if attempts is not None:
    customers = customer_summary(attempts)
    print(f"Unique customers: {customers['customer_id'].nunique():,}")
    print(customers["drop_off"].value_counts(normalize=True).rename("share"))
```

```python
if attempts is not None:
    failures = attempts.loc[attempts["status"].eq("fail"), "stop_point"]
    ax = failures.value_counts().head(10).sort_values().plot.barh(figsize=(9, 5))
    ax.set(title="Highest-volume onboarding failure points", xlabel="Failed attempts", ylabel="")
    plt.tight_layout()
```

### Reported funnel findings

- 9,148 unique customers were observed.
- 93% successfully completed onboarding.
- More than 600 customers dropped off.
- OCR and OTP accounted for approximately 80% of failures.
- The proposal estimated approximately VND 4.8 billion in acquisition value at risk; this is an estimate, not realized accounting loss.


## 5. Leakage-Safe Feature Engineering

Historical features must be computed before the current event. The use of `shift()` below prevents a customer's present outcome from entering its own predictors.


```python
def engineer_features(df: pd.DataFrame) -> pd.DataFrame:
    x = df.sort_values(["customer_id", "begin_time"]).copy()
    x["hour_of_day"] = x["begin_time"].dt.hour
    x["day_of_week"] = x["begin_time"].dt.dayofweek
    x["is_weekend"] = x["day_of_week"].ge(5).astype(int)
    x["is_office_hour"] = (
        x["day_of_week"].lt(5) & x["hour_of_day"].between(8, 17)
    ).astype(int)
    x["attempt_number"] = x.groupby("customer_id").cumcount() + 1

    prior_fail = x["status"].eq("fail").astype(int)
    x["prior_failures"] = prior_fail.groupby(x["customer_id"]).transform(
        lambda s: s.shift(fill_value=0).cumsum()
    )
    prior_pending = x["status"].eq("pending").astype(int)
    x["prior_pending"] = prior_pending.groupby(x["customer_id"]).transform(
        lambda s: s.shift(fill_value=0).cumsum()
    )
    return x

featured_attempts = engineer_features(attempts) if attempts is not None else None
```

## 6. Customer-Level Modeling Dataset

The model used the final observed record for each customer, enriched only with information available before or at that attempt. Categorical preprocessing was fitted inside each cross-validation fold.


```python
FEATURES = [
    "gender", "phone_type", "device_system_name", "os_version",
    "hour_of_day", "day_of_week", "is_weekend", "is_office_hour",
    "attempt_number", "prior_failures", "prior_pending",
]

if featured_attempts is not None:
    model_df = customer_summary(featured_attempts)
    available_features = [c for c in FEATURES if c in model_df.columns]
    X = model_df[available_features]
    y = model_df["drop_off"]
    categorical = X.select_dtypes(include=["object", "category"]).columns.tolist()
    numeric = [c for c in available_features if c not in categorical]
```

## 7. Model Comparison

The competition solution compared XGBoost, LightGBM, and Random Forest with stratified cross-validation. ROC-AUC was primary because the bank needed a useful risk ranking; accuracy alone is misleading when drop-off is uncommon.


```python
reported_results = pd.DataFrame(
    {
        "ROC-AUC": [0.8943, 0.8927, 0.8907],
        "Accuracy": [0.9287, 0.9303, 0.9351],
        "F1": [0.5462, 0.5442, 0.5255],
        "Precision": [0.4742, 0.4808, 0.5100],
        "Recall": [0.6463, 0.6281, 0.5438],
    },
    index=["XGBoost", "LightGBM", "Random Forest"],
)
reported_results.style.highlight_max(axis=0, color="#bfdbfe").format("{:.4f}")
```

```python
# Reproducible baseline that runs when authorized data is available.
if featured_attempts is not None:
    preprocessor = ColumnTransformer(
        transformers=[
            ("categorical", OneHotEncoder(handle_unknown="ignore"), categorical),
            ("numeric", "passthrough", numeric),
        ]
    )
    baseline = Pipeline(
        steps=[
            ("preprocess", preprocessor),
            ("model", RandomForestClassifier(
                n_estimators=400,
                class_weight="balanced",
                random_state=RANDOM_STATE,
                n_jobs=-1,
            )),
        ]
    )
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
    scores = cross_validate(
        baseline, X, y, cv=cv,
        scoring=["roc_auc", "accuracy", "f1", "precision", "recall"],
        n_jobs=-1,
    )
    pd.DataFrame(scores).filter(like="test_").mean()
```

## 10. Step-Specific Prediction: A Useful Negative Result

Separate models were tested for OTP, OCR, privacy, selfie, contract, and other failure steps. Their ROC-AUC values generally remained around 0.5-0.6, close to weak discrimination. These models were not presented as deployment-ready.

Their feature importance was used only as exploratory context. Final recommendations relied more strongly on funnel volume, cohort behavior, domain analysis, and the substantially stronger overall drop-off model.


## 11. Business Actions

The evidence supported six priorities:

1. Improve OCR capture guidance and recovery.
2. Strengthen OTP delivery, resend, and error messaging.
3. Detect OS, camera, and NFC incompatibility before a blocking step.
4. Add accessible guidance and assisted support for older customers.
5. Preserve progress so customers can resume instead of restart.
6. Validate changes with funnel instrumentation and controlled experiments.

See [`docs/business-recommendations.md`](../docs/business-recommendations.md) for the measurement framework.


## 12. Limitations And Production Readiness

- The analysis covered only Q1 2024.
- Observational associations are not causal effects.
- Some timestamp and device fields required quality treatment.
- The financial impact used an external customer acquisition cost assumption.
- Production deployment would require temporal validation, calibration, fairness review, drift monitoring, and prospective A/B testing.

The reported 93% to above-97% completion improvement was a proposal target to test, not a guaranteed forecast.
